### Step 3. Run the pre-processing pipeline and demonstrate STFT result

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
from scipy import signal

def plot_spec_heatmap(spec, t, win_len):
    # Plot the heatmap
    plt.figure(figsize=(10, 10), dpi=300)
    plt.rcParams.update({'font.size': 20})
    plt.tight_layout()

    # Create heatmap
    sea.heatmap(spec.squeeze()[1:, :], cbar=False)

    # Set X-axis (Time) labels
    start_index = min(preamble_2_idx)  # Starting point of your IQ samples
    time_ticks = np.linspace(0, len(t)-1, 5).astype(int)  # Select 5 evenly spaced time bins
    plt.xticks(ticks=time_ticks, labels=(start_index + np.round(t[time_ticks] * 25)).astype(int))  # Adjust time labels

    # Set Y-axis (Frequency) labels
    freq_bins = np.fft.fftshift(np.fft.fftfreq(win_len))  # Frequency indices
    freq_ticks = np.linspace(0, len(freq_bins)-1, 10).astype(int)  # Select 7 evenly spaced frequency bins
    plt.yticks(ticks=freq_ticks, labels=np.round(freq_bins[freq_ticks] * win_len).astype(int))

    # Adjust the padding for the tick labels to avoid overlap
    plt.gca().tick_params(axis='x', pad=10)  # Increase padding for X-axis ticks
    plt.gca().tick_params(axis='y', pad=10)  # Increase padding for Y-axis ticks

    # Set axis labels
    plt.xlabel('Time (samples)')
    plt.ylabel('Frequency (index)')

    # Display the plot
    plt.show()

# Your data and preprocessing steps
data = np.array([preamble_1])  # Adjust 'preamble_1' to your actual data

# Normalization (assuming it's part of your code)
data = ChannelIndSpectrogram()._normalization(data)

win_len = 50
overlap = 25

# STFT computation
f, t, spec = signal.stft(data, 
                         window='boxcar', 
                         nperseg=win_len, 
                         noverlap=overlap, 
                         nfft=win_len,
                         return_onesided=False, 
                         padded=False, 
                         boundary=None)

# FFT shift to adjust the central frequency
spec = np.fft.fftshift(spec, axes=0)

# Take the logarithm of the magnitude.      
spec = abs(spec)

plot_spec_heatmap(spec, t, win_len=50)
```

### Step 4. Obtain channel-independent spectrogram & plot it as well

```python
# Generate channel independent spectrogram.
chan_ind_spec = spec[:,1:]/spec[:,:-1]    

# Take the logarithm of the magnitude.      
chan_ind_spec = np.log10(np.abs(chan_ind_spec)**2)

plot_spec_heatmap(chan_ind_spec, t, win_len=49)
```
